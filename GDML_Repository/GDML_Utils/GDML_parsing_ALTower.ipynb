## GDML Parsing
This notebook can be used to change the GDML file defining:
- Material of the components
- Properly define the Sensitive detectors
- Set the colors

### Search the file name for the Parsing

```python
import os
import sys


DirectoryGDML = "/home/riccardo/Documenti/GeantProjects/LEM_GDML_upgrade/GDML_Repository/ALTower_2"


Actual_dir = os.getcwd()
os.chdir(DirectoryGDML)
GlobalPath = os.getcwd()
os.chdir(Actual_dir)

# Trova il file con estensione .gdml

for file in os.listdir(DirectoryGDML):
    if file.endswith(".gdml") and not "Parsed" in file:
        print(file)
        # Get the name with the global path 
        
        FileNameInput = DirectoryGDML + file
        FileNameInput = os.path.join(GlobalPath, file)
        print(FileNameInput)
```

```python
FileInputGDML = open(FileNameInput, "r")

VolumeName = []
VolumeNameref = []

# Fai un loop su tutte le righe del file:
# Se incontri la stringa "volume name=" allora salva il nome del volume (racchiuso tra le virgolette "") nel vettore VolumeName
# Se incontri la stringa "volumeref ref=" allora salva il nome del volume (racchiuso tra le virgolette "") nel vettore VolumeNameref

for line in FileInputGDML:
    if "volume name=" in line:
        VolumeName.append(line.split('"')[1])
    if "volumeref ref=" in line:
        VolumeNameref.append(line.split('"')[1])

FileInputGDML.close()
```

```python
# Stampa in due file distinti le due liste di nomi appena trovate

FileOutputVolumeName = open(os.path.join(DirectoryGDML,"VolumeName.txt"), "w")
for i in VolumeName:
    FileOutputVolumeName.write(i+"\n")

FileOutputVolumeName.close()

FileOutputVolumeNameRef = open(os.path.join(DirectoryGDML,"VolumeNameRef.txt"), "w")
for i in VolumeNameref:
    FileOutputVolumeNameRef.write(i+"\n")

FileOutputVolumeNameRef.close()
```

```python
# Duplica i files in modo da poter modificare la seconda versione

import shutil

if os.path.isfile(os.path.join(DirectoryGDML,"VolumeName2.txt")):
    print('Stai attento: Il file esiste già')
    # Errore
else:
    shutil.copy(os.path.join(DirectoryGDML,"VolumeName.txt"), os.path.join(DirectoryGDML,"VolumeName2.txt"))
    print('Il file è stato copiato con successo')

if os.path.isfile(os.path.join(DirectoryGDML,"VolumeNameRef2.txt")):
    print('Stai attento: Il file esiste già')
else:
    shutil.copy(os.path.join(DirectoryGDML,"VolumeNameRef.txt"), os.path.join(DirectoryGDML,"VolumeNameRef2.txt"))
    print('Il file è stato copiato con successo')

#!cp ../GDML_LEM-19032023/VolumeName.txt ../GDML_LEM-19032023/VolumeName2.txt
#!cp ../GDML_LEM-19032023/VolumeNameRef.txt ../GDML_LEM-19032023/VolumeNameRef2.txt
```

# Modifica i nomi degli oggetti nel file numerato con il numero 2

```python
# Apri VolumeNameRef2.txt e anche VolumeName.txt

VolumeName2 = []
VolumeNameRef2 = []

File2 = open(os.path.join(DirectoryGDML, "VolumeName2.txt"), "r")
for line in File2:
    VolumeName2.append(line[:-1])

File2.close()

FileRef2 = open(os.path.join(DirectoryGDML, "VolumeNameRef2.txt"), "r")
for line in FileRef2:
    VolumeNameRef2.append(line[:-1])

FileRef2.close()

print(VolumeName2)
print(VolumeNameRef2)
```

## Controlla le differenze

```python
# Trova le differenze tra i due vettori VolumeName e VolumeName2

Counter = 0
for index , name in enumerate(VolumeName):
    if not name in VolumeName2:
        print(str(index) + " " + name + " => "+ VolumeName2[index])
        Counter += 1

print("Changes: " + str(Counter))


print("\n")
Counter = 0
# Trova le differenze tra i due vettori VolumeNameRef e VolumeNameRef2
for index , name in enumerate(VolumeNameref):
    if not name in VolumeNameRef2:
        print(str(index) + " " + name + " => "+ VolumeNameRef2[index])
        Counter += 1

print("Changes: " + str(Counter))
```

## Inizio il parsing vero e proprio

```python
GAGGdefinition = '''

<materials>
  <element name="Gd" formula="Gd" Z="64.">
    <atom value="157.25"/>
      </element>
  <element name="Al" formula="Al" Z="13.">
    <atom value="26.981539"/>
      </element>
  <element name="Ga" formula="Ga" Z="31.">
    <atom value="69.723"/>
      </element>
  <element name="O" formula="O" Z="8.">
    <atom value="16.00"/>
      </element>


  <material name="GAGG" formula="Gd3Al2Ga3O12">
    <D value="6.63" unit="g/cm3"/>  
    <composite n="3" ref="Gd"/>
    <composite n="2" ref="Al"/>
    <composite n="3" ref="Ga"/>
    <composite n="12" ref="O"/>
  </material>
  
</materials>

'''
print(GAGGdefinition)
```

```python
# Togli .gdml dal nome del file 
FileNameParsed = FileNameInput[:-5] + "_Parsed.gdml"
#FileNameParsed = DirectoryGDML+"GDML_file_Parsed.gdml"


DummyMaterials = True
IsSilicon = False
IsPlastic = False

FileReport = open(os.path.join(DirectoryGDML, "Report.txt"), "w")

FileOutputGDML = open(FileNameParsed, "w")
FileInputGDML = open(FileNameInput, "r")


#FileOutputGDML.write(GAGGdefinition)


for line in FileInputGDML:
    SetSensDet = False

    if "volume name=" in line:
        print("##############################################")
        FileReport.write("##############################################")
        print("Found volume name in line\n" + line) 
        FileReport.write("Found volume name in line\n" + line)
        for i in range(len(VolumeName)):
            if VolumeName[i] in line:
                print("Volume name recognized :" + VolumeName[i])
                FileReport.write("Volume name recognized :" + VolumeName[i])
                line = line.replace(VolumeName[i], VolumeName2[i])
                lineName = line
                print("Replacing with :" + VolumeName2[i])
                FileReport.write("Replacing with :" + VolumeName2[i])
                print(line)
                FileReport.write(line)
                FileOutputGDML.write(line)
                break
            
    elif "volumeref ref=" in line:
        print("##############################################")
        FileReport.write("##############################################")
        print("Found volumeref ref in line\n" + line)
        FileReport.write("Found volumeref ref in line\n" + line)
        for i in range(len(VolumeNameref)):
            if VolumeNameref[i] in line:
                print("Volume name recognized :" + VolumeNameref[i])
                FileReport.write("Volume name recognized :" + VolumeNameref[i])
                line = line.replace(VolumeNameref[i], VolumeNameRef2[i])
                lineName = line
                print("Replacing with :" + VolumeNameRef2[i])
                FileReport.write("Replacing with :" + VolumeNameRef2[i])
                print(line)
                FileReport.write(line)
                FileOutputGDML.write(line)
                break
            
    elif "materialref" in line:
        print("Found materialref in line\n" + line)
        FileReport.write("Found materialref in line\n" + line)
        IsPlastic = False
        IsSilicon = False
        if "LV_Plastic" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_PLASTIC_SC_VINYLTOLUENE\"/>\n"
            IsPlastic = True
        elif "LV_Al" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_Al\"/>\n"
        elif "LV_ALTAI" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_Si\"/>\n"
            IsSilicon = True
        elif "worldVOL" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_Galactic\"/>\n"
        elif "LV_Kapton" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_Si\"/>\n"
        elif "LV_COMPOUND" in lineName:
            lineMat = "\t\t\t<materialref ref=\"G4_Al\"/>\n"
        else:
            pass
        
        if (not IsSilicon) and (not IsPlastic) and DummyMaterials:
            lineMat = "\t\t\t<materialref ref=\"G4_Galactic\"/>\n"
            print("Dummy Materials ON:\n Not Plastic and Not Silicon are replaced with Vacuum")
            FileReport.write("Dummy Materials ON:\n Not Plastic and Not Silicon are replaced with Vacuum")
            FileReport.write("Replacing with :\n" + lineMat)
            
            
            
            
        print("Replacing with :\n" + lineMat)
        FileReport.write("Replacing with :\n" + lineMat)
        FileOutputGDML.write(lineMat)
    elif "auxiliary auxtype=\"Color\"" in line:
        print("Found auxiliary auxtype=\"Color\" in line\n" + line)
        FileReport.write("Found auxiliary auxtype=\"Color\" in line\n" + line)
        if "LV_Plastic" in lineName: #Light blue in RGBA HEX like before
            lineCol = "\t\t\t<auxiliary auxtype=\"Color\" auxvalue=\"#ffffffff\"/>\n"
            print("Setting color to light blue")
        if "LV_Al" in lineName: #Grey in RGBA HEX like before
            lineCol = "\t\t\t<auxiliary auxtype=\"Color\" auxvalue=\"#ff808080\"/>\n"
            print("Setting color to grey")
        if "LV_ALTAI" in lineName: #Green in RGBA HEX like before
            lineCol = "\t\t\t<auxiliary auxtype=\"Color\" auxvalue=\"#ff00ff00\"/>\n"
            print("Setting color to green")
        if "LV_Kapton" in lineName: #Pink in RGBA HEX like before
            lineCol = "\t\t\t<auxiliary auxtype=\"Color\" auxvalue=\"#ffff00ff\"/>\n"
            print("Setting color to pink")
        if "shield" in lineName: #Orange in RGBA HEX like before
            lineCol = "\t\t\t<auxiliary auxtype=\"Color\" auxvalue=\"#ffffa500\"/>\n"
            print("Setting color to orange")
        print(lineCol)
        FileReport.write(lineCol)
        FileOutputGDML.write(lineCol)

        if "LV_ALTAI" in lineName:
            lineSensDet = "\t\t\t<auxiliary auxtype=\"SensDet\" auxvalue=\"Silicon\"/>\n"
            SetSensDet = True
        if "LV_Plastic" in lineName:
            lineSensDet = "\t\t\t<auxiliary auxtype=\"SensDet\" auxvalue=\"PlasticScintillator\"/>\n"
            SetSensDet = True
        
        if SetSensDet == True:
            print("SensDet found")
            FileReport.write("SensDet found")
            FileOutputGDML.write(lineSensDet)
            FileReport.write(lineSensDet)
            SetSensDet = False  
    #elif "<materials/>" in line:
    #    FileOutputGDML.write(GAGGdefinition)
    
    else:    
        FileOutputGDML.write(line)

FileOutputGDML.close()
FileInputGDML.close()
```
