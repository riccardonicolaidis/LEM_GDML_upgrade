## Creazione e riempimento di un TTree
Numeri distribuiti secondo una distribuzione gaussiana in XY con Box Muller

```python
tree = rt.TTree("tree", "tree")
x = np.zeros(1, dtype=float)
y = np.zeros(1, dtype=float)

X = np.zeros(1, dtype=float)
Y = np.zeros(1, dtype=float)

N = 1e6

tree.Branch("X", X, "X/D")
tree.Branch("Y", Y, "Y/D")

gPad = rt.gPad

for i in range(int(N)):
    x[0] = np.random.uniform(0, 1)
    y[0] = np.random.uniform(0, 1)
    X[0] = np.sin(2 * np.pi * x[0]) * np.sqrt(-2 * np.log(y[0]))
    Y[0] = np.cos(2 * np.pi * x[0]) * np.sqrt(-2 * np.log(y[0]))
    tree.Fill()
    
c = rt.TCanvas("c", "c", 800, 600)
h = rt.TH2D("h", "h", 200, -4, 4, 200, -4, 4)
h.SetMinimum(0)

tree.Draw("Y:X>>h", "", "colz")
#h.Smooth()
gPad.SetLogz()
c.Draw()
```

## Tentativo con TGraph

```python
tree2 = rt.TTree("tree2", "tree2")

t = np.zeros(1, dtype=float)
s = np.zeros(1, dtype=float)
v = np.zeros(1, dtype=float)

tree2.Branch("t", t, "t/D")
tree2.Branch("s", s, "s/D")
tree2.Branch("v", v, "v/D")

N = 1e5

Fs = 5e3
dt = 1/Fs
f1 = 13
f2 = 19


omega1 = 2 * np.pi * f1
omega2 = 2 * np.pi * f2

for i in range(int(N)):
    t[0] = i * dt
    s[0] = 5 * np.sin(omega1 * t[0]) + 0.01 * np.random.normal(0, 1)
    v[0] = 5 * np.sin(omega2 * t[0]) + 0.01* np.random.normal(0, 1)
    tree2.Fill()

c2 = rt.TCanvas()
tree2.Draw("v:s", "", "goff")
g = rt.TGraph(tree2.GetSelectedRows(), tree2.GetV2(), tree2.GetV1())
g.Draw("al")
g.SetTitle("Lissajous curve")
g.SetMarkerStyle(8)
g.SetMarkerSize(0.5)
g.SetLineColor(rt.kRed)

gPad.SetGrid()

c2.Draw()
```
