Test della funzione os.walk()

```python
import os
import shutil


TextDir = os.path.join(os.getcwd(), "TextDir")

shutil.rmtree(TextDir)


if not os.path.exists(TextDir):
    os.makedirs(TextDir)
    
#Create subfolders with a depth of N
N = 3


def CreateSubfolders(StartingPath, NumberSubfolders, RecursiveDepth):
    if RecursiveDepth > 0:
        for i in range(NumberSubfolders[0]):
            FolderName = "Subfolder_"+str(RecursiveDepth)+"_"+str(i)
            os.makedirs(os.path.join(StartingPath, FolderName))
            CreateSubfolders(os.path.join(StartingPath, FolderName), NumberSubfolders[1:], RecursiveDepth-1)
        else:
            f1 = open(os.path.join(StartingPath, "file1.txt"), "w")
            f1.write("This is a test file")
            f1.close()
            
            
CreateSubfolders(TextDir, [5,3,4], 3)


for root, dirs, files in os.walk(TextDir):
    print(root)
    print(dirs)
    print(files)
    print("\n\n")
```
